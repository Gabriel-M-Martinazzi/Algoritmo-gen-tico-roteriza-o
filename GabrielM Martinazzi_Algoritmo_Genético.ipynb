1 - Para ler o arquivo:

```python
import re

with open('xqf131.tsp') as f:
    lines = f.readlines()

# Encontra a linha que começa com "NODE_COORD_SECTION"
node_coord_section_line = next((line for line in lines if line.startswith('NODE_COORD_SECTION')), None)

# Encontra o índice da linha que começa com "NODE_COORD_SECTION"
node_coord_section_line_index = lines.index(node_coord_section_line)

# Extrai as coordenadas das cidades
city_coords = []
for line in lines[node_coord_section_line_index + 1:]:
    if line.strip() == 'EOF':
        break
    city_id, x, y = re.split(r'\s+', line.strip())
    city_coords.append((float(x), float(y)))

# Imprime as coordenadas das cidades
for i, (x, y) in enumerate(city_coords):
    print(f'Cidade {i + 1}: ({x}, {y})')
```

3 - Criando uma população de 10 indivíduos

```python
import random

# Cria uma lista de 10 indivíduos, onde cada indivíduo é uma lista de cidades
population = []
for i in range(10):
    # Embaralha as coordenadas das cidades para criar uma rota aleatória
    route = random.sample(city_coords, len(city_coords))
    population.append(route)

# Imprime a população
for i, individual in enumerate(population):
    print(f'Indivíduo {i + 1}: {individual}')

```

4 - Calculando a fitness

```python
import math

def distance(city1, city2):
    x1, y1 = city1
    x2, y2 = city2
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def route_distance(route):
    total_distance = 0
    for i in range(len(route)):
        city1 = route[i]
        city2 = route[(i + 1) % len(route)]
        total_distance += distance(city1, city2)
    return total_distance

# Calcula a fitness de cada indivíduo na população
fitness_scores = []
for individual in population:
    fitness_scores.append(route_distance(individual))

# Imprime as fitness scores
for i, score in enumerate(fitness_scores):
    print(f'Indivíduo {i + 1}: {score}')
```

5 - Utilizando o método de seleção "Torneio"

```python
import random

# Função para seleção por torneio
def selecao_por_torneio(populacao, aptidoes, tamanho_torneio):
    pais_selecionados = []
    for _ in range(len(populacao)):
        torneio = random.sample(range(len(populacao)), tamanho_torneio)
        vencedor_torneio = max(torneio, key=lambda x: aptidoes[x])
        pais_selecionados.append(populacao[vencedor_torneio])
    return pais_selecionados

# Tamanho do torneio (quantos indivíduos competirão)
tamanho_torneio = 5

# Realizar a seleção por torneio
pais_selecionados = selecao_por_torneio(population, fitness_scores, tamanho_torneio)

# Imprimir os pais selecionados
for i, pai in enumerate(pais_selecionados):
    print(f"Pai {i + 1}: {pai}")
```
